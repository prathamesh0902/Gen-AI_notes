```sql
-- Prompting and Prompt Templates
Prompt - instructions provided to an LLM to guide its response.
Chain of Thought prompting / RGC prompting

Prompt components - Role, Goal, Context, Output Format, Constraints/Consistency.

Role - defines who the model should act as   ::  Data Scientist, Teacher, Recruiter.

Goal - defines the task to be completed.

Context - provides background information required for the task.

Output Format - specifies expected response structure   ::  JSON, Table, Bullet Points.

Consistency/Constraints - rules the model must follow.
→ Well-structured prompts produce more reliable outputs.

Static Prompting - fixed prompt with hardcoded values.    -   Can lead to halluciantion      :: 8 lines of code
Dynamic Prompting - prompt generated using variables and placeholders.

PromptTemplate - reusable prompt structure with dynamic placeholders.    -  can be asved as JSON

Example placeholders - {role}, {goal}, {query}.

Why Prompt Templates?
- Reusability.
- Dynamic user inputs.
- Cleaner code.
- Easier maintenance.

Template validation - verifies all required variables are provided.
Missing parameter validation - raises errors when placeholders are not supplied.
validate_template=True - checks template correctness before execution.

PromptTemplate output - formatted prompt string ready for the LLM.

Flow:
Template + Variables → Formatted Prompt → LLM → Output Parser


Multi-model Prompting - output from one model becomes input to another model.

Example:
Prompt → Gemini → GPT → Parser

→ Multiple models can be combined to leverage different strengths.

-- Messages and ChatPrompot Template
Chat Models - operate using messages instead of a single prompt string.

Message Types:
SystemMessage - defines behavior and instructions.
HumanMessage - contains user query.
AIMessage - contains model response.
SystemMessage - usually stores role, goals, constraints, and context.
HumanMessage - provides runtime user input.
AIMessage - stores previous model outputs.

Chat conversation:
SystemMessage → HumanMessage → AIMessage

→ Messages maintain conversational structure and context.


ChatPromptTemplate - template for creating multi-message conversations.

PromptTemplate - best for single text prompts.

ChatPromptTemplate - best for chat-based interactions.

Dynamic Chat Prompting - inject variables into system and human messages at runtime.

Example:
SystemMessage = "You are a {role}"
HumanMessage = "{question}"

→ Variables are assigned when the chain executes.


Conversation Memory - stores previous messages to maintain context across interactions.

Multi-message prompting - uses current and historical messages together.

Dynamic multi-message prompting - roles, context, and queries can all be assigned at runtime.


LangChain Imports:

langchain_core.prompts
- PromptTemplate
- ChatPromptTemplate

langchain_core.messages
- SystemMessage
- HumanMessage
- AIMessage

langchain_community.llms
- Community-supported LLM integrations such as Ollama and other open-source models.


Prompt Workflow:
PromptTemplate/ChatPromptTemplate → LLM → Output Parser → Final Response
```

```sql
-- Structured Output from LLM and Pydantic
LLMs can generate unstructured text such as employee details in paragraph format.
Example - "There are 25 employees in this department..." (free-form text)

Unstructured output - difficult to send directly to APIs, databases, or downstream applications.

Structured output - data returned in predefined formats such as JSON, dictionaries, tables, or key-value pairs.
Example - Name, City, Age.

Only prompting is often insufficient when applications need reliable machine-readable outputs.
→ Structured outputs enable direct integration with databases, APIs, and AI agents.

Data extraction - convert LLM responses into usable structured fields.
Example - Job portal resume extraction.

API building - extract product reviews, ratings, positive/negative sentiment from customer feedback.
Example - Amazon review analysis.

Agent building - AI agents use structured outputs to interact with tools and external systems.

Tool calling - LLM generates structured parameters required by a tool or API.
→ Structured outputs form the bridge between LLM applications and real-world actions.

OpenAI, Anthropic, Gemini - support native structured output generation.

Hugging Face and Ollama models - may not provide native structured outputs.
Output parsers - convert raw text responses into structured formats.

Pydantic - Python data validation and serialization library used for structured outputs.
Pydantic model - defines expected schema, fields, and data types.

Type safety - ensures values match expected data types.   ::   Age must be integer, Name must be string.

Validation - verifies both data types and business rules.     ::   Age > 0.

Field constraints - define requirements, defaults, optional values, descriptions, and validations.

Example Resume Schema:
Name - string
Age - integer
City - string

Pydantic object - structured response returned from the LLM.
Dictionary conversion - response.model_dump() or response.dict().
JSON conversion - response.json().

Structured responses can be passed to APIs, databases, AI agents, and HTTP requests.
→ Structured data becomes the common communication format between systems.

Chains - connect multiple LLMs sequentially.
Example - LLM 1 generates summary → LLM 2 generates final 5-point response.

Output parsers - transform raw LLM text into JSON, CSV, dictionaries, or custom schemas.
Output parsers ensure consistency and validation when native structured outputs are unavailable.

StrOutputParser - extracts only plain text from an LLM response.

LLM responses often contain content plus metadata.
StrOutputParser - returns only the content text and removes metadata.

Example Chain:
Prompt/topic → LLM 1 detailed report → LLM 1 summary → LLM 2 final response.

Flow:
Prompt → LLM Response → Output Parser → Structured Data → API/Database/Agent
```

```sql
-- Output Parsers in Langchain and StrOutputParser
Data ingestion - collecting data from external sources before processing.
Data cleaning - removing errors, duplicates, and inconsistencies from data.
Data preprocessing - transforming raw data into model-ready format.
→ Clean and preprocess data before sending it to LLM workflows.

Chain - sequence of Prompt → LLM → Parser.   ::   Prompt → Gemini → Prompt → GPT → Output Parser.


-- JSON and Pydantic Output Parser
JSONOutputParser - converts LLM responses into JSON format.
PydanticOutputParser - converts responses into validated Pydantic objects.
Output parser - enforces predefined response schema.

Format instructions - parser-generated instructions added to prompts to guide output structure.
PromptTemplate - reusable prompt with placeholders for dynamic values.
Partial variables - predefined variables automatically injected into prompts.
Pydantic validation - catches missing fields, incorrect types, and invalid values before downstream processing.

-- Parallel and Conditional Chains
Chain - combines multiple LangChain components into a workflow.
Sequential Chain - output of one step becomes input to the next step.
Parallel Chain - multiple chains execute simultaneously.
Conditional Chain - execution path changes based on conditions or previous outputs.
→ Different chain types help build complex AI workflows.

-- Runnables and LCEL
Runnable - a unit of work that accepts input, processes it, and produces output.

All LangChain components are runnables.
Examples - Prompt Templates, LLMs, Retrievers, Parsers.

invoke() - executes a runnable on a single input.
batch() - executes a runnable on multiple inputs.
stream() - returns output incrementally as it is generated.

Runnables can be connected together to create end-to-end workflows.
→ Prompt → Retriever → LLM → Parser.

RAG Pipeline:
Document Loader - loads documents.
Text Splitter - breaks documents into chunks.
Embedding Model - converts text into vectors.
Vector Store - stores embeddings.
Retriever - fetches relevant chunks.
LLM - generates final response.

Runnable primitives - fundamental building blocks of LangChain.
RunnableSequence - executes steps sequentially.
RunnableParallel - executes multiple runnables concurrently.
RunnableBranch - routes execution based on conditions.
RunnablePassthrough - forwards input unchanged.
RunnableLambda - wraps custom Python functions.

Task-specific runnables - specialized runnables for retrievers, LLMs, prompts, and parsers.

LangChain Core - collection of standardized runnable components.
→ Complex AI workflows are built by composing simple runnables.


LPU (Language Processing Unit) - specialized processor optimized for fast LLM inference.
LPU - designed for running trained models, not training them.
Groq LPU - custom chip architecture focused on ultra-low latency inference.
Inference - generating predictions or responses using a trained model.
GPU - general-purpose parallel processor widely used for AI training and inference.
NVIDIA GPUs - industry standard hardware for training large AI models.
TPU (Tensor Processing Unit) - Google\s specialized hardware for large-scale AI workloads.
TPUs - optimized for training models on massive datasets.

Hardware Flow:
TPU/GPU → Model Training
LPU/GPU → Model Inference
```
