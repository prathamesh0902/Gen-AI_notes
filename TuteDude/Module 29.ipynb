| Component             | Purpose                                                                                      | Importance                                                                         |
| --------------------- | -------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------- |
| **Data Source**       | PDFs, Word files, websites, databases, YouTube transcripts, etc.                             | Provides the knowledge base that the LLM can retrieve from.                        |
| **Document Loader**   | Loads documents into memory. Examples: `PyPDFLoader`, `WebBaseLoader`.                       | Converts raw files into LangChain document objects.                                |
| **Text Splitter**     | Breaks large documents into smaller chunks. Example: `RecursiveCharacterTextSplitter`.       | LLMs and embedding models have token limits; chunking improves retrieval accuracy. |
| **Chunks**            | Smaller pieces of text generated by splitting.                                               | Retrieval happens at chunk level, not entire document level.                       |
| **Embedding Model**   | Converts text chunks into numerical vectors. Example: `all-MiniLM-L6-v2`, OpenAI Embeddings. | Captures semantic meaning of text for similarity search.                           |
| **Embeddings**        | Dense vector representations of text. Example: `[0.12, -0.45, 0.78, ...]`.                   | Allows mathematical comparison of text similarity.                                 |
| **Vector Store**      | Stores embeddings and enables fast similarity search. Examples: FAISS, ChromaDB, Pinecone.   | Core retrieval engine of RAG.                                                      |
| **Retriever**         | Searches vector store and returns top-k relevant chunks.                                     | Finds context relevant to the user's query.                                        |
| **User Query**        | Question asked by the user.                                                                  | Input that drives retrieval.                                                       |
| **Query Embedding**   | Converts user query into a vector.                                                           | Enables similarity comparison against stored document vectors.                     |
| **Similarity Search** | Finds nearest chunk vectors to query vector using cosine similarity or Euclidean distance.   | Retrieves most relevant information.                                               |
| **Context Window**    | Retrieved chunks combined together.                                                          | Provides external knowledge to the LLM.                                            |
| **Prompt Template**   | Combines context and user query into a structured prompt.                                    | Guides the LLM to answer based on retrieved data.                                  |
| **LLM**               | Generates final response. Examples: Llama 3, GPT-4, Gemma.                                   | Produces human-readable answer.                                                    |
| **Response**          | Final answer shown to the user.                                                              | End result of the RAG pipeline.                                                    |


Shifting from Fine Tuning to RAG

| Feature                       | Fine-Tuning                            | In-Context Learning (ICL)                         | RAG (Retrieval Augmented Generation)                  |
| ----------------------------- | -------------------------------------- | ------------------------------------------------- | ----------------------------------------------------- |
| **Definition**                | Retrains model weights on custom data. | Provides examples and instructions in the prompt. | Retrieves external knowledge and feeds it to the LLM. |
| **Knowledge Storage**         | Stored permanently in model weights.   | Stored temporarily in prompt context.             | Stored externally in vector database/documents.       |
| **Model Weights Change?**     | ✅ Yes                                  | ❌ No                                              | ❌ No                                                  |
| **Training Required?**        | ✅ Yes                                  | ❌ No                                              | ❌ No                                                  |
| **Access to New Information** | Requires retraining.                   | Only if included in prompt.                       | Simply update documents/vector store.                 |
| **Cost**                      | High                                   | Low                                               | Medium                                                |
| **Implementation Complexity** | High                                   | Very Low                                          | Medium                                                |
| **Inference Speed**           | Fast                                   | Fast                                              | Slightly slower (retrieval step added).               |
| **Scalability**               | Poor for frequent updates.             | Limited by context window.                        | Excellent for large knowledge bases.                  |
| **Hallucination Reduction**   | Moderate                               | Low                                               | High                                                  |
| **Best For**                  | Learning specialized behavior/style.   | Few-shot learning and task guidance.              | Knowledge-intensive applications.                     |


```sql
LLM - trained on massive amounts of internet text data.

LLM Pipeline:
Prompt → Embeddings → LLM → Response

RAG Pipeline:
Document Loader → Text Splitter → Embeddings → Vector Store → Retriever → LLM

Machine Learning - learns patterns by fitting relationships from input (X) to output (Y).
Example - Y = MX + C.

Deep Learning - neural networks with multiple hidden layers that learn complex patterns.

Backpropagation - updates model weights by propagating prediction errors backward through the network.
→ Deep learning powers modern LLMs.

LLM knowledge is stored inside weights and biases.

Parameters - numerical values representing learned knowledge.

Model Size:
70B Parameters > 13B Parameters > 7B Parameters

Larger models generally have greater reasoning and knowledge capacity.


Limitations of Base LLMs:
- No access to personal/private data.
- No access to current real-time data.
- Can hallucinate incorrect answers.
- Knowledge limited to training cutoff date.

Hallucination - generating confident but incorrect information.


Pretraining - training on large-scale text using next-word prediction.

Next-word prediction - model learns to predict the most likely next token.

Fine-Tuning - retraining a pretrained model on domain-specific data.

Supervised Fine-Tuning (SFT) - training with labeled input-output examples.
→ Fine-tuning modifies model behavior by updating weights.


Full Parameter Fine-Tuning - updates all model parameters.

Challenges:
- Expensive training.
- Requires large datasets.
- Requires GPUs.
- Requires AI/ML expertise.
- Risk of memorizing sensitive data.
- Privacy concerns.
- Retraining required for new information.

Dynamic data cannot be updated daily through fine-tuning.

Fine-tuning changes model knowledge permanently.


Parameter-Efficient Fine-Tuning (PEFT) - updates only a small subset of parameters.

Frozen Weights - original model parameters remain unchanged.

Trainable Parameters - only selected layers or adapters are updated.

Benefits:
- Lower cost.
- Faster training.
- Less GPU memory.
- Preserves original model knowledge.


In-Context Learning (ICL) - teaches the model through examples inside the prompt without updating weights.

Example:
Text: "I love this product."
Sentiment: Positive

Text: "The service was terrible."
Sentiment: Negative

Text: "{new_text}"
Sentiment: ?

→ Model learns from examples provided in the prompt.

Zero-Shot Prompting - task instruction only, no examples.

Few-Shot Prompting - task instruction plus a few examples.

Advantages:
- No retraining.
- No GPU requirements.
- No technical expertise.

Limitations:
- Limited by context window.
- Prompt size increases token cost.
- Information disappears after the conversation ends.

Prompting does not modify model weights.


Context Window - maximum tokens processed by the model.

Example:
GPT-3.5 ≈ 16K Tokens

Pricing - based on input tokens + output tokens.

After the context window is exceeded, older information is forgotten.


RAG (Retrieval-Augmented Generation) - provides external knowledge during inference without retraining.

Retriever - fetches relevant documents for the user/s query.

Vector Store - stores document embeddings for efficient retrieval.

Retrieved documents are added as context before calling the LLM.

Flow:
User Query → Retriever → Relevant Documents → Prompt + Context → LLM Response

Augmentation - adding retrieved context to the prompt.

Formula:
Augmentation = Prompt + Retrieved Context


Benefits of RAG:
- Uses private/company data.
- Access to current information.
- Lower cost than fine-tuning.
- No model retraining.
- Better data privacy.
- Easier to maintain.
- Lower computational requirements.

RAG retrieves only relevant documents instead of sending the entire dataset.

Typical Flow:
Query → Top 3-5 Relevant Documents → LLM

Comparison:

Prompting
- Uses instructions/examples only.
- No external knowledge.

Fine-Tuning
- Updates model weights.
- Permanent knowledge change.
- Expensive.

RAG
- Retrieves external documents.
- No weight updates.
- Dynamic and cost-effective.
```

**"Large" Language Models**

```sql
LLMs are called "Large" because of the scale of their parameters, training data, and computational resources compared to earlier language models.

Large Number of Parameters

Parameters are the weights the model learns during training. More parameters → higher capacity to learn complex patterns in language.
Modern LLMs have billions to trillions of parameters.
Example: GPT‑2 → 1.5B, GPT‑3 → 175B, GPT‑4 → even more.

Large Training Data

Trained on massive text corpora from books, articles, websites, code, etc.
Often hundreds of gigabytes to terabytes of text.
This scale helps capture diverse knowledge and linguistic nuances.


Large Computational Requirements

Training requires thousands of GPUs/TPUs running for weeks or months.
Memory and compute needs are far beyond typical machine learning models.


Large Context & Capability

Can process long sequences of text (thousands of tokens).
Can perform a wide range of tasks without task-specific training (zero-shot/few-shot learning).
```
