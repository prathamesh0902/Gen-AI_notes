Module 31 - Buildnig Chatbots with Converstaion History

```sql
-- Building Chatbots with Message History
Memory - mechanism for preserving conversation context across multiple interactions.

Stateless APIs - each request is independent and does not remember previous interactions.

Memory component - adds statefulness by storing and retrieving chat history.
→ Memory enables multi-turn conversations.

BaseChatMessageHistory - abstract base class for chat history implementations.

Core methods:
add_messages() - stores messages.
get_messages() - retrieves messages.

ChatMessageHistory - simplest memory implementation using an in-memory Python list.

Message History Stores:
ChatMessageHistory - local memory.
RedisChatMessageHistory - Redis-backed persistent storage.
FirestoreChatMessageHistory - Google Firestore-backed storage.

RunnableWithMessageHistory - wraps a chain and automatically manages chat history.


Normal Chain:
Prompt → Model → Parser

Memory-enabled Chain:
Prompt + History → Model → Parser
→ Previous conversations are injected into every new request.

Session ID - unique identifier for a conversation.
get_session_history() - retrieves history associated with a session ID.
Different session IDs maintain separate conversations.

Example:
Session A → own history.
Session B → own history.


Input Messages Key - identifies where current user messages are passed.
History Messages Key - identifies where stored conversation history is injected.
Output Messages Key - identifies which responses should be stored back into memory.

Flow:
User Query → History Injection → LLM → Response → Save Response

-- Messages Placeholder
ChatPromptTemplate with Memory:
System Message - instructions.
History Placeholder - previous messages.
Human Message - current query.

Flow:
History + Current Query → Prompt → LLM → Response


Context Window - maximum number of tokens an LLM can process at once.

Common Context Lengths:
4K Tokens
8K Tokens
32K Tokens
100K+ Tokens (some modern models)

If chat history grows indefinitely, token limits will eventually be exceeded.
→ Old history must be trimmed or compressed.

-- Trimming Chat history
Trimming Strategy - removes older messages when context becomes too large.

Message Count Trimming - keep only the last N messages.
Example - last 10 interactions.

Token Count Trimming - keep messages until token budget is reached.

Sliding Window Strategy - continuously remove oldest messages first.

Example:
Keep latest 2000 tokens.
Discard older messages.


Conversation Summarization - compress old conversations into a short summary.

Example:
50K-token history → summarized to 1K-token memory.

Summary Memory - stores key facts while reducing token usage.

Memory Management Flow:
Chat History → Trim/Summarize → Inject into Prompt → Generate Response


Complete Flow:
User Query + Session ID
→ Retrieve History
→ Insert History into Prompt
→ LLM Response
→ Store New Messages
→ Trim/Summarize if Needed
→ Updated Memory
```
