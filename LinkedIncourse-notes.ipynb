NLP has evolved from early neural language models and word embeddings like Word2Vec, which capture semantic meaning but can also reflect biases from training data.
Recurrent neural networks (RNNs) and sequence-to-sequence models improved NLP by handling variable-length input and output sequences, but struggled with long-term dependencies.
The introduction of attention mechanisms in 2015 allowed models to focus on relevant parts of the input sequence dynamically, overcoming limitations of fixed-length encoding.
The 2017 "Attention is All You Need" paper introduced the transformer architecture, which uses attention as the primary mechanism, enabling more efficient and powerful language models like BERT and GPT.

Self-attention is a mechanism used in transformer models where each word in a sentence relates to every other word in that same sentence. This allows the model to understand how words connect to each other, like how pronouns refer to other words or how different parts of a sentence relate grammatically. For example, in the sentence "My friend told me about this class and I love it," self-attention helps the model recognize that "I" refers to "me" and "she" refers to "friend." This ability to capture relationships within the sentence is what makes transformers so powerful for natural language processing tasks.

Transformers differ from recurrent neural networks (RNNs) in how they handle sequence to sequence modeling. While RNNs compress the entire input sequence into a single fixed-length vector before decoding, transformers use an encoder-decoder architecture that relies on self-attention instead of recursion. This means transformers process the input into a rich matrix representation through multiple stacked encoders, capturing relationships between all tokens simultaneously. The decoder then uses cross-attention to connect this input representation with the output sequence it generates. This approach allows transformers to handle long sequences more efficiently and capture complex dependencies better than RNNs, making them state-of-the-art for many NLP tasks.

The transformer architecture uses an encoder-decoder setup to handle sequence-to-sequence tasks. The encoder is a stack of multiple encoders that process the input sequence (like a sentence) and convert it into a rich matrix representation instead of a single fixed vector. This allows the model to capture complex relationships between all words simultaneously.

The decoder is also a stack of decoders that take this encoded representation and generate the output sequence step-by-step. It uses a special mechanism called cross-attention to connect the input representation from the encoder with the output it produces, enabling it to predict the next word in the sequence effectively.

Additionally, the encoder uses multi-headed attention to understand different grammatical structures at once, while the decoder uses masked multi-headed attention to generate outputs in a coherent, stepwise manner. This architecture allows transformers to efficiently model long sequences and complex dependencies, making them powerful for NLP tasks like translation or text generation.

### Definitions:
- An encoder is a neural network component that takes an input sequence (text, image, audio, etc.) and maps it into a numerical representation (embedding or feature vector) that captures its meaning or structure.
- A decoder is a neural network component that takes the encoder’s output (or its own previous outputs) and generates a target sequence step-by-step.
- A Transformer is a deep learning architecture introduced in "Attention Is All You Need" (Vaswani et al., 2017) that uses self-attention instead of recurrence or convolution to process sequences.

---

1. Encoder
Input → hidden representation.
Often uses self-attention (in Transformers) or RNN/CNN layers in older architectures.
Output is a contextual embedding for each input token (or a single vector for the whole input).

Example in NLP:

In BERT, the encoder takes a sentence and outputs contextual embeddings for each word.
In machine translation, the encoder reads the source sentence and produces a representation for the decoder.


2. Decoder
Uses attention to focus on relevant parts of the encoder’s output.
Can be auto-regressive (predicts one token at a time, feeding it back in).
In Transformers, the decoder has masked self-attention so it can’t see future tokens.

Example in NLP:

In translation, the decoder takes the encoder’s representation of a French sentence and generates the English translation.
In GPT-like models, the decoder generates text without an encoder (decoder-only architecture).


3. Transformer
Encoder-Decoder structure in the original paper:

Encoder: multiple layers of self-attention + feed-forward networks.
Decoder: multiple layers of masked self-attention + encoder-decoder attention + feed-forward networks.


Self-attention lets each token attend to all others in the sequence.
Highly parallelizable → faster training than RNNs.
Variants:

Encoder-only: BERT (good for classification, embeddings).
Decoder-only: GPT (good for text generation).
Encoder-Decoder: T5, original Transformer (good for translation, summarization).




Diagram (Conceptual)
Input Sequence → [Encoder] → Encoded Representation → [Decoder] → Output Sequence

Transformer internals:
Encoder Layer:  [Self-Attention] → [Feed Forward] → Output
Decoder Layer:  [Masked Self-Attention] → [Encoder-Decoder Attention] → [Feed Forward] → Output

Scaled Dot-Product Attention is a core mechanism in transformer models that helps the model focus on different parts of the input sequence when processing language. Here's how it works:

The input tokens are first transformed into three matrices: query (Q), key (K), and value (V). Each represents the tokens in different vector spaces with specific meanings—Q represents what we're looking for, K represents the relevance of tokens to the query, and V holds the semantic meaning of tokens without context.

The attention scores are calculated by multiplying the query matrix by the transpose of the key matrix. This results in a matrix showing how much each token should pay attention to every other token.

These scores are then scaled down by dividing by the square root of the embedding dimension to keep the values manageable and speed up computation.

A softmax function is applied to convert these scores into probabilities that sum to one, indicating the attention distribution.

Finally, these attention weights are multiplied by the value matrix to produce context-aware representations of each token, meaning the model now understands each word in relation to the others in the sequence.


This process allows transformers to capture complex relationships in text efficiently, making them powerful for NLP tasks. The video you are watching goes into detail with examples and code to help solidify this concept.

-----

Multi-headed attention is a technique used in transformer models where instead of having just one attention mechanism, the model has several smaller attention mechanisms (called heads) running in parallel within each encoder layer. Each head focuses on different parts or aspects of the input text, like different grammatical relationships or word connections. For example, one head might focus on pronoun references while another focuses on direct object relations.

These multiple attention heads work independently, and their outputs are combined by concatenating them and then transforming them into a single, richer representation. This allows the model to capture multiple language patterns simultaneously, making it better at understanding the context and meaning of words in a sentence. In BERT-base, for instance, there are 12 encoder layers each with 12 attention heads, totaling 144 attention calculations happening at once.

This multi-headed approach helps transformers like BERT to grasp complex relationships in language, improving their performance on various NLP tasks.

Transfer learning in transformers means starting with a model that’s already been trained on a large amount of general language data, like BERT trained on Wikipedia and BookCorpus. Instead of training a model from scratch, you take this pre-trained model and fine-tune it on a smaller, specific dataset for a particular task, such as classification or question answering. This approach leverages the model’s existing understanding of language, making it faster and more effective to adapt to new tasks.

There are three common fine-tuning strategies:

- Update all layers of the model including the new task-specific layers (best performance but slower training).
- Freeze some of the earlier layers and only update the later layers plus new layers (balances speed and performance).
- Freeze the entire pre-trained model and only train the new layers on top (fastest but may perform worse).

This method is powerful because it transfers the knowledge learned from large, general datasets to your specific NLP problems, saving time and improving results.

BERT transforms contextless embeddings into contextful representations using the scaled dot-product attention mechanism. Initially, each token has a contextless embedding (the value matrix V) that captures its meaning without considering other words. BERT calculates attention scores by multiplying the query (Q) and key (K) matrices, scaling them, and applying a softmax to get probabilities showing how much each token should focus on others. These attention scores are then multiplied by the value matrix V, which adjusts each token's embedding based on the context from the entire sequence. This process lets BERT create rich, context-aware representations of words, essential for understanding language in NLP tasks.

## Multi headed attention

In multi-headed attention, each head has its own distinct set of query (Q), key (K), and value (V) matrices. This means that for different heads, all three matrices—Q, K, and V—differ. Having separate Q, K, and V matrices per head allows each head to focus on different parts or aspects of the input, enabling the transformer to capture a richer variety of relationships in the data.

In transformer models, especially in the attention mechanism, Q, K, and V stand for Query, Key, and Value matrices:

Query (Q): Represents the information you're looking for in the input—think of it as the 'question' or focus point.
Key (K): Represents how relevant each token (word or piece of text) is to the query—it's like a measure of importance or connection.
Value (V): Contains the actual semantic meaning of each token without considering context—basically, the raw information.

The attention mechanism uses these matrices to determine how much focus each token should get relative to others, enabling the model to create context-aware representations of the input text.

# Cross Attention

```python
Attention - mechanism that allows a model to focus on the most relevant parts of the input.

Self-Attention - Query (Q), Key (K), and Value (V) are generated from the same sequence.

Example:
Sentence → Self-Attention → Each word attends to other words in the same sentence.

Cross-Attention - Query (Q) comes from one sequence while Key (K) and Value (V) come from another sequence.

Formula:
Attention(Q,K,V) = softmax(QKᵀ/√d)V

Cross-Attention:
Q → Target Sequence
K,V → Source Sequence

→ Allows one sequence to retrieve information from another sequence.

Example:
English Input → "I love AI"
French Output → "J'aime l'IA"

While generating "J'aime", the decoder attends to relevant English words using cross-attention.

Transformer Encoder-Decoder:
Encoder → Produces contextual embeddings.
Decoder Self-Attention → Looks at previously generated words.
Decoder Cross-Attention → Looks at encoder outputs.

Flow:
Input Text → Encoder
Generated Tokens → Decoder Self-Attention
Encoder Output + Decoder Output → Cross-Attention
→ Next Token Prediction

Purpose of Cross-Attention:
- Connects two different information sources.
- Retrieves relevant context.
- Aligns source and target sequences.

Applications:
Machine Translation - Source sentence ↔ Target sentence.
RAG - User query ↔ Retrieved documents.
Image Captioning - Text tokens ↔ Image features.
Vision-Language Models - Text ↔ Image embeddings.

RAG Example:
Query - "What is LangChain?"
Retrieved Documents - LangChain documentation.

Cross-Attention - query attends to relevant document chunks before generating the answer.

Difference:

Self-Attention
- Q, K, V from same sequence.
- Learns relationships within a sequence.

Cross-Attention
- Q from one sequence.
- K, V from another sequence.
- Learns relationships across sequences.

Memory Trick:
Self-Attention = "Look within."
Cross-Attention = "Look outside."
```
